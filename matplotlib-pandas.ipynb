```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
data={
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [10000, 12000, 15000, 13000, 17000, 16000],
    "Profit": [2000, 3000, 4000, 2500, 3500, 3000]
}

df= pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Month'], df['Sales'], color='blue', marker='o', linestyle='-', label='Sales')
plt.title('Sales Trend Over Months')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
width=0.3
plt.bar(df['Month'], df['Sales'], width=width, color='skyblue', label='Sales')
plt.title('Sales trends over months')
plt.xlabel('Moths')
plt.ylabel('Sales($)')
plt.grid(True)
plt.legend()
```

```python
plt.figure(figsize=(10, 6))
width = 0.3
plt.bar(df['Month'], df['Sales'], width=width, label='Sales', color='skyblue')
plt.bar(df['Month'], df['Profit'], width=width, label='Profit', color='orange', bottom=df['Sales'])
plt.title('Sales and Profit Comparison by Month')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.pie(df['Profit'],labels=df['Month'],autopct='%1.1f%%',startangle=140,colors=plt.cm.Paired.colors)
plt.title('profit by months')
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Sales'], df['Profit'], color='green', s=100, edgecolors='black')
plt.title('Sales vs Profit')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.hist(df['Sales'], bins=5, color='purple', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot(df['Profit'], vert=False, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.title('Profit Distribution')
plt.xlabel('Profit ($)')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up a grid of subplots
fig, axs = plt.subplots(5, 3, figsize=(15, 20))
fig.suptitle('Common Types of Graphs in Matplotlib', fontsize=18)
axs = axs.flatten()

x = np.linspace(0, 10, 100)
y = np.sin(x)

# 1. Line Plot
axs[0].plot(x, y)
axs[0].set_title("Line Plot")

# 2. Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
axs[1].bar(categories, values)
axs[1].set_title("Bar Chart")

# 3. Scatter Plot
x_rand = np.random.rand(50)
y_rand = np.random.rand(50)
axs[2].scatter(x_rand, y_rand)
axs[2].set_title("Scatter Plot")

# 4. Histogram
data = np.random.randn(1000)
axs[3].hist(data, bins=30)
axs[3].set_title("Histogram")

# 5. Pie Chart
axs[4].pie(values, labels=categories, autopct='%1.1f%%')
axs[4].set_title("Pie Chart")

# 6. Box Plot
data2 = [np.random.randn(100) for _ in range(4)]
axs[5].boxplot(data2)
axs[5].set_title("Box Plot")

# 7. Stack Plot
x_stack = np.arange(1, 6)
y1 = np.random.rand(5)
y2 = np.random.rand(5)
axs[6].stackplot(x_stack, y1, y2, labels=['y1', 'y2'])
axs[6].legend(loc='upper left')
axs[6].set_title("Stack Plot")

# 8. Error Bar Plot
yerr = 0.2
axs[7].errorbar(x, y, yerr=yerr)
axs[7].set_title("Error Bar Plot")

# 9. Heatmap
matrix = np.random.rand(10, 10)
axs[8].imshow(matrix, cmap='viridis')
axs[8].set_title("Heatmap")

# 10. Contour Plot
X, Y = np.meshgrid(x, x)
Z = np.sin(X) * np.cos(Y)
axs[9].contourf(X, Y, Z, cmap='coolwarm')
axs[9].set_title("Contour Plot")

# 11. Stem Plot
axs[10].stem(x[:10], y[:10])
axs[10].set_title("Stem Plot")

# 12. Quiver Plot
Xq, Yq = np.meshgrid(np.arange(-2, 2, 0.5), np.arange(-2, 2, 0.5))
U = -Yq
V = Xq
axs[11].quiver(Xq, Yq, U, V)
axs[11].set_title("Quiver Plot")

# 13. Placeholder for 3D Plot
axs[12].text(0.5, 0.5, '3D Plot\n(Needs mpl_toolkits)', ha='center', va='center', fontsize=10)
axs[12].set_title("3D Plot")
axs[12].axis('off')

# 14. Polar Plot
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(5*theta))
axs[13] = plt.subplot(5, 3, 14, polar=True)
axs[13].plot(theta, r)
axs[13].set_title("Polar Plot")

# 15. Step Plot
axs[14].step(x, y)
axs[14].set_title("Step Plot")

plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```
